```python
import astropy.units as u
from astropy.coordinates.sky_coordinate import SkyCoord
from astropy.units import Quantity
from astroquery.gaia import Gaia
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

# Suppress warnings. Comment this out if you wish to see the warning messages
import warnings
warnings.filterwarnings('ignore')
```

```python
# Looking for the Andromeda Galaxy
job = Gaia.launch_job_async("SELECT * \
FROM gaiadr1.gaia_source \
WHERE CONTAINS(POINT('ICRS',gaiadr1.gaia_source.ra,gaiadr1.gaia_source.dec),CIRCLE('ICRS',10.675,41.2692,2))=1;" \
                            , dump_to_file=True)
print (job)
```

```python
r = job.get_results()
print (r['source_id'])
```

```python
plt.scatter(r['ra'], r['dec'], color='b', alpha=0.1)
plt.xlim(10,10.7)
plt.ylim(42,42.7)

plt.show()
```
